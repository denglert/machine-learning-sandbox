# Cross entropy

For continuous case:

\begin{equation}
    H
    =
    - \int p(x) \log q(x) dx
\end{equation}

For discrete case:

\begin{equation}
    H
    =
    - \sum_{i}
    p(x_{i}) \log q(x_{i})
\end{equation}

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def normalize(w):
    return w/sum(w)
```

```python
def cross_entropy(p, q):
    h = - np.dot(p, np.log(q))
    return h
```

```python
def plot_distributions(pdf1, pdf2):
    fig, ax = plt.subplots()
    x = np.arange(len(pdf1))
    ax.bar(x-0.15, pdf1, color='firebrick', width=0.3, edgecolor='k', label='1', alpha=0.5)
    ax.bar(x+0.15, pdf2, color='navy', width=0.3, edgecolor='k', label='2', alpha=0.5)
    ax.legend()
    h = cross_entropy(pdf1, pdf2)
    ax.set_title("Cross entropy: {:.3f}".format(h))
    return fig, ax
```

```python
def create_dashboard(w1, w2):
    w1 = np.array(w1)
    w2 = np.array(w2)
    pdf1 = normalize(w1)
    pdf2 = normalize(w2)
    fig, ax = plot_distributions(pdf1, pdf2)
```
